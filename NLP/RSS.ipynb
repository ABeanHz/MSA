# Use RSS to pull articles from El Pais

```python
def run_rss(topic, cat, news):  
    # Use requests to fetch the feed
    import requests
    response = requests.get(topic)

    # Parse the feed
    feed = feedparser.parse(response.content)
    print(feed)


    # Extract the article information
    articles = []

    for entry in feed.entries:
        article = {
            "title": entry.get("title", ""),
            "link": entry.get("link", ""),
            "published": entry.get("published", ""),
            "summary": entry.get("summary", ""),
        }
        articles.append(article)



    # Convert to a dataframe
    df = pd.DataFrame(articles)

    df['category'] = cat
    df['periodico'] = news

    # Check the number of articles
    print(f"Number of articles: {len(df)}")
    
    return(df)
```

```python
# List of RSS feed URLs and their corresponding categories
rss_addresses = [
    ("https://feeds.elpais.com/mrss-s/pages/ep/site/elpais.com/section/internacional/portada", "international", "elpais"),
    ("https://feeds.elpais.com/mrss-s/pages/ep/site/elpais.com/section/sociedad/portada", "sociedad", "elpais"), 
    ("https://feeds.elpais.com/mrss-s/pages/ep/site/elpais.com/section/economia/portada", "economia", "elpais"), 
    ("https://feeds.elpais.com/mrss-s/pages/ep/site/elpais.com/section/ciencia/portada", "ciencia", "elpais"), 
    ("https://feeds.elpais.com/mrss-s/pages/ep/site/elpais.com/section/deportes/portada", "deportes", "elpais"), 
    ("https://feeds.elpais.com/mrss-s/list/ep/site/elpais.com/section/clima-y-medio-ambiente", "clima", "elpais"), 
    ("https://feeds.elpais.com/mrss-s/list/ep/site/elpais.com/section/economia/subsection/negocios", "negocios", "elpais"), 
    ("https://feeds.elpais.com/mrss-s/list/ep/site/elpais.com/section/gastronomia", "gastronomia", "elpais"), 
    ("https://feeds.elpais.com/mrss-s/list/ep/site/elpais.com/section/educacion", "educacion", "elpais"), 
    ("https://feeds.elpais.com/mrss-s/list/ep/site/elpais.com/section/elviajero", "viajes", "elpais")
]

# Initialize an empty list to store the DataFrames
dfs = []

# Loop through each RSS feed and process it
for rss_url, category, newspaper in rss_addresses:
    df = run_rss(rss_url, category, newspaper)
    dfs.append(df)

# Concat into 1 dataframe
final_df = pd.concat(dfs, ignore_index=True)

# see final df
print(final_df)
```
