# Read in libraries

```python
import feedparser
import pandas as pd
```

# Use RSS to pull articles from spanish language news sources

```python
def run_rss(topic, cat, news):  
    # Use requests to fetch the feed
    import requests
    response = requests.get(topic)

    # Parse the feed
    feed = feedparser.parse(response.content)
    print(feed)


    # Extract the article information
    articles = []

    for entry in feed.entries:
        article = {
            "title": entry.get("title", ""),
            "link": entry.get("link", ""),
            "published": entry.get("published", ""),
            "summary": entry.get("summary", ""),
        }
        articles.append(article)



    # Convert to a dataframe
    df = pd.DataFrame(articles)

    df['category'] = cat
    df['periodico'] = news

    # Check the number of articles
    print(f"Number of articles: {len(df)}")
    
    return(df)
```
